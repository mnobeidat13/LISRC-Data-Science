```python
import numpy as np
import matplotlib.pyplot as plt

def plot_math_function(expr, x_range=(-10, 10), title="Math Function Plot"):
    # Generate integer x values within the specified range
    x = np.arange(x_range[0], x_range[1] + 1)
    
    # Evaluate the expression to calculate y values
    y = [eval(expr) for x in x]
    
    # Print x and y values
    print(f'x: {", ".join(map(str, x))}')
    print(f'y: {", ".join(map(str, y))}')
    
    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'y = {expr}')
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    
    # Show the plot without the grid
    plt.show()

# Example usage:
# Pass the math function as a string
math_expr = "2*x +2"
plot_math_function(math_expr, x_range=(0, 10), title="Plot of y = 2x + 1")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data points scattered around a mean
np.random.seed(0)
mean = 30
std_dev = 5
num_points = 20
data_points = np.random.normal(mean, std_dev, num_points)

# Create a horizontal line representing the mean
mean_line = np.full(num_points, mean)

# Calculate two horizontal lines representing mean + std and mean - std
std_line_upper = np.full(num_points, mean + std_dev)
std_line_lower = np.full(num_points, mean - std_dev)

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(range(num_points), data_points, label='Data Points', c='b', marker='o', alpha=0.5)
plt.plot(mean_line, label=f'Mean ({mean})', linestyle='--', color='r')
plt.plot(std_line_upper, label=f'Mean + Std ({mean + std_dev})', linestyle='--', color='g')
plt.plot(std_line_lower, label=f'Mean - Std ({mean - std_dev})', linestyle='--', color='y')

# Set the y-axis limit to start from zero
plt.ylim(0, max(data_points) + std_dev)

# Add vertical lines from each data point to the mean
for i in range(num_points):
    plt.vlines(x=i, ymin=data_points[i], ymax=mean, linestyle='-', color='purple', alpha=0.5)

plt.title('Scatter Plot with Mean and Standard Deviation Lines')
plt.xlabel('Data Points')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Whole Class with additions:
class Neural_Network(object):
    def __init__(self):        
        #Define Hyperparameters
        self.inputLayerSize = 2
        self.outputLayerSize = 1
        self.hiddenLayerSize = 3
        
        #Weights (parameters)
        self.W1 = np.random.randn(self.inputLayerSize,self.hiddenLayerSize)
        self.W2 = np.random.randn(self.hiddenLayerSize,self.outputLayerSize)
        
    def forward(self, X):
        #Propogate inputs though network
        self.z2 = np.dot(X, self.W1)
        self.a2 = self.sigmoid(self.z2)
        self.z3 = np.dot(self.a2, self.W2)
        yHat = self.sigmoid(self.z3) 
        return yHat
        
    def sigmoid(self, z):
        #Apply sigmoid activation function to scalar, vector, or matrix
        return 1/(1+np.exp(-z))
    
    def sigmoidPrime(self,z):
        #Gradient of sigmoid
        return np.exp(-z)/((1+np.exp(-z))**2)
    
    def costFunction(self, X, y):
        #Compute cost for given X,y, use weights already stored in class.
        self.yHat = self.forward(X)
        J = 0.5*sum((y-self.yHat)**2)
        return J
        
    def costFunctionPrime(self, X, y):
        #Compute derivative with respect to W and W2 for a given X and y:
        self.yHat = self.forward(X)
        
        delta3 = np.multiply(-(y-self.yHat), self.sigmoidPrime(self.z3))
        dJdW2 = np.dot(self.a2.T, delta3)
        
        delta2 = np.dot(delta3, self.W2.T)*self.sigmoidPrime(self.z2)
        dJdW1 = np.dot(X.T, delta2)  
        
        return dJdW1, dJdW2
```
